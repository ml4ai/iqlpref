```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from prettytable import PrettyTable

pd.options.mode.copy_on_write = True
```

```python
mr_cloned_reduce_0 = pd.read_csv(
    "../cloned_pen_models_pref/iql-p-D4RL/pen/stats_reduce_0.csv",
    header=None,
    names=["return"],
)

mr_cloned_reduce_0["reduction"] = 0

mr_cloned_reduce_10 = pd.read_csv(
    "../cloned_pen_models_pref/iql-p-D4RL/pen/stats_reduce_10.csv",
    header=None,
    names=["return"],
)

mr_cloned_reduce_10["reduction"] = 10

mr_cloned_reduce_20 = pd.read_csv(
    "../cloned_pen_models_pref/iql-p-D4RL/pen/stats_reduce_20.csv",
    header=None,
    names=["return"],
)

mr_cloned_reduce_20["reduction"] = 20

mr_cloned_reduce_30 = pd.read_csv(
    "../cloned_pen_models_pref/iql-p-D4RL/pen/stats_reduce_30.csv",
    header=None,
    names=["return"],
)

mr_cloned_reduce_30["reduction"] = 30

mr_cloned_reduce_40 = pd.read_csv(
    "../cloned_pen_models_pref/iql-p-D4RL/pen/stats_reduce_40.csv",
    header=None,
    names=["return"],
)

mr_cloned_reduce_40["reduction"] = 40

mr_cloned_reduce_50 = pd.read_csv(
    "../cloned_pen_models_pref/iql-p-D4RL/pen/stats_reduce_50.csv",
    header=None,
    names=["return"],
)

mr_cloned_reduce_50["reduction"] = 50

mr_cloned_reduce_60 = pd.read_csv(
    "../cloned_pen_models_pref/iql-p-D4RL/pen/stats_reduce_60.csv",
    header=None,
    names=["return"],
)

mr_cloned_reduce_60["reduction"] = 60

mr_cloned_reduce_70 = pd.read_csv(
    "../cloned_pen_models_pref/iql-p-D4RL/pen/stats_reduce_70.csv",
    header=None,
    names=["return"],
)

mr_cloned_reduce_70["reduction"] = 70

mr_cloned_reduce_80 = pd.read_csv(
    "../cloned_pen_models_pref/iql-p-D4RL/pen/stats_reduce_80.csv",
    header=None,
    names=["return"],
)

mr_cloned_reduce_80["reduction"] = 80

mr_cloned_reduce_90 = pd.read_csv(
    "../cloned_pen_models_pref/iql-p-D4RL/pen/stats_reduce_90.csv",
    header=None,
    names=["return"],
)

mr_cloned_reduce_90["reduction"] = 90

mr_cloned_df = pd.concat(
    [
        mr_cloned_reduce_0,
        mr_cloned_reduce_10,
        mr_cloned_reduce_20,
        mr_cloned_reduce_30,
        mr_cloned_reduce_40,
        mr_cloned_reduce_50,
        mr_cloned_reduce_60,
        mr_cloned_reduce_70,
        mr_cloned_reduce_80,
        mr_cloned_reduce_90,
    ]
)
```

```python
cloned_means = mr_cloned_df.groupby("reduction")["return"].mean().reset_index()
cloned_stds = mr_cloned_df.groupby("reduction")["return"].std().reset_index()
cloned_counts = mr_cloned_df.groupby("reduction")["return"].count().reset_index()
```
