```python
tr_data = {
    "Dataset": datasets,
    "Average Normalized Score": nms_tr,
    "Std Normalized Score": nstd_tr,  # Precomputed error for each category
}
tr_df = pd.DataFrame(tr_data)
tr_df["Reward Function"] = "Task Reward"

mr_data = {
    "Dataset": datasets,
    "Average Normalized Score": nms_mr,
    "Std Normalized Score": nstd_mr,  # Precomputed error for each category
}
mr_df = pd.DataFrame(mr_data)
mr_df["Reward Function"] = "MLP"

pt_data = {
    "Dataset": datasets,
    "Average Normalized Score": nms_pt,
    "Std Normalized Score": nstd_pt,  # Precomputed error for each category
}
pt_df = pd.DataFrame(pt_data)
pt_df["Reward Function"] = "PT"

br_fg_data = {
    "Dataset": datasets,
    "Average Normalized Score": nms_br_fg,
    "Std Normalized Score": nstd_br_fg,  # Precomputed error for each category
}
br_fg_df = pd.DataFrame(br_fg_data)
br_fg_df["Reward Function"] = "BNN (FG)"

br_optim_data = {
    "Dataset": datasets,
    "Average Normalized Score": nms_br_optim,
    "Std Normalized Score": nstd_br_optim,  # Precomputed error for each category
}
br_optim_df = pd.DataFrame(br_optim_data)
br_optim_df["Reward Function"] = "BNN (optim)"

br_optim_star_data = {
    "Dataset": datasets,
    "Average Normalized Score": nms_br_optim_star,
    "Std Normalized Score": nstd_br_optim_star,  # Precomputed error for each category
}
br_optim_star_df = pd.DataFrame(br_optim_star_data)
br_optim_star_df["Reward Function"] = "BNN (optim_star)"

df = pd.concat([tr_df, mr_df, pt_df, br_fg_df, br_optim_df, br_optim_star_df])

# Create the Seaborn barplot without default error bars
r_ax = sns.barplot(
    x="Dataset",
    y="Average Normalized Score",
    hue="Reward Function",
    data=df,
    errorbar=None,
)

# Add custom error bars using Matplotlib
# Get the x-positions of the bars for errorbar placement
n_rfs = len(df["Reward Function"].unique())
bar_width = 0.8 / n_rfs
offsets = np.linspace(-bar_width * (n_rfs - 1) / 2, bar_width * (n_rfs - 1) / 2, n_rfs)
datasets = r_ax.get_xticks()
rf_order = df["Reward Function"].unique()

for i, dataset_label in enumerate(df["Dataset"].unique()):
    for j, rf_label in enumerate(rf_order):
        subset = df[
            (df["Dataset"] == dataset_label) & (df["Reward Function"] == rf_label)
        ]
        if not subset.empty:
            x_pos = datasets[i] + offsets[j]
            y_val = subset["Average Normalized Score"].iloc[0]
            y_err = subset["Std Normalized Score"].iloc[0]
            r_ax.errorbar(
                x_pos, y_val, yerr=y_err, fmt="none", color="black", capsize=5
            )

plt.xticks(rotation=30)
plt.title("Scores")
plt.show()
```
